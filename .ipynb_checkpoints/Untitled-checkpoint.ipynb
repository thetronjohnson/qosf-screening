### Implement a circuit that returns |01> and |10> with equal probability

```python
from qiskit import *
import numpy as np
from qiskit.visualization import plot_histogram
from qiskit.aqua.operators
%matplotlib inline
```

```python
theta,phi = 0,0
backend = Aer.get_backend('qasm_simulator')
rate = 0.01
def circuit(p):
    circ = QuantumCircuit(1)
    circ.ry(p,0)
    return circ
```

```python
while(1):
    circ = circuit(theta)
    circ.draw('mpl')
    circ.measure_all()
    job = execute(circ,backend=backend,shots=100)
    counts = job.result().get_counts()
    try:
        pr_1 = counts['1']
    except:
        theta += 0.1
    
    fn = ((np.tan(theta/2))**2 - 1)
    cost = fn**2
    
    if cost<0.001:
        print(f"Optimal theta: {theta}")
        break
    dcost = 2*(fn)*np.tan(theta/2)/(np.cos(theta/2)**2)
    theta = (theta - rate*dcost)%(np.pi*2)
```

We know that the q0 needs to have an expectation value of 0 and q1 a value of -1 wrt sigma Z

```python
while(1):
    circ = circuit(phi)
    circ.draw('mpl')
    circ.measure_all()
    job = execute(circ,backend=backend,shots=100)
    counts = job.result().get_counts()
    try:
        pr_1 = counts['1']
    except:
        phi += 0.1
    
    fn = ((np.tan(phi/2))**2)
    cost = fn**2
    
    if cost<0.001:
        print(f"Optimal phi: {phi}")
        break
    dcost = 2*(fn)*np.tan(phi/2)/(np.cos(phi/2)**2)
    phi = (phi - rate*dcost)%(np.pi*2)
```

```python
qc = QuantumCircuit(2)
qc.ry(theta,0)
qc.ry(phi,1)
qc.cx(0,1)
qc.measure_all()
qc.draw('mpl')
job = execute(qc,backend=backend,shots=100)
counts = job.result().get_counts()
```

```python
plot_histogram(counts)
```

```python
theta,phi = 0,0
backend = Aer.get_backend('qasm_simulator')
rate = 0.01
def circuit(t,p):
    circ = QuantumCircuit(2)
    circ.ry(t,0)
    circ.ry(p,1)
    circ.cx(0,1)
    return circ
```

```python
while(1):
    circ = circuit(theta,phi)
    circ.draw('mpl')
    circ.measure_all()
    job = execute(circ,backend=backend,shots=100)
    counts = job.result().get_counts()
    try:
        pr_01 = counts['01']
    except:
        phi += 0.1
    try:
        pr_01 = counts['10']
    except:
        theta += 0.1
    
    tan_t = np.tan(theta/2)
    tan_p = np.tan(phi/2)
    sec_t = 1/np.cos(theta/2)
    sec_p = 1/np.cos(phi/2)
    fn = ((tan_t/tan_p)**2 -1 )
    cost = fn**2
    
    
    if cost<0.00000000000000000001:
        print(f"Optimal theta: {theta}")
        print(f"Optimal phi: {phi}")
        break
    dcost = 2*fn*2*(tan_t/tan_p)*((tan_p*(sec_t**2))-(tan_t*(sec_p**2))/tan_p**2)
    theta -= (rate*dcost)%(np.pi*2)
    phi += (rate*dcost)%(np.pi*2)
```

```python
qc = QuantumCircuit(2)
qc.ry(theta,0)
qc.ry(phi,1)
qc.measure_all()
qc.draw('mpl')
job = execute(qc,backend=backend,shots=100)
counts = job.result().get_counts()
plot_histogram(counts)
```
